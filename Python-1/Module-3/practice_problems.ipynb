```python
# Problem 1:

with open ("example.txt", "r") as file:
    data = file.readlines()

    # Total lines:
    total_lines = len(data)
    print("Total lines:", total_lines)

    # Number of words:
    number_of_words = list(map(lambda x : len(x.split()),data))
    print("Number of words:", number_of_words)

    # Total words:
    from functools import reduce 
    total_words = reduce(lambda x,y : x+y, number_of_words)
    print("Total words:", total_words)

    # Using strip() method to remove new lines:
    data = list(map(lambda x : x.strip(),data))
    print(data)

    # Detele spaces:
    data = list(map(lambda x : x.replace(" ", ""),data))
    print(data)

    # Character count:
    char = list(map(lambda x : len(x),data))
    print("Character count:", char)

    # Total character:
    total_char = reduce(lambda x,y : x+y, char)
    print("Total character:", total_char)

    # Save all data to other file:
    with open ("counter_of_string.txt","w") as file:
        file.write("Total lines: " + str(total_lines) + "\n")
        file.write("Number of words: " + str(number_of_words) + "\n")
        file.write("Total words: " + str(total_words) + "\n")
        file.write("Character count: " + str(char) + "\n")
        file.write("Total character: " + str(total_char) + "\n")
```

```python
# File Pointer: It will tell us where our curson is in the file.

with open ("example.txt", "r") as file:
    print(file.tell())
    print(file.read(5))
    print(file.tell())
    print(file.read())
    print(file.tell())
```

```python
# Using seek() method how we change the position of the file pointer:

with open ("example.txt", "r") as file:
    print(file.tell())
    print(file.read(5))
    print(file.tell())
    file.seek(0)
    print(file.tell())
    print(file.read(5))
```

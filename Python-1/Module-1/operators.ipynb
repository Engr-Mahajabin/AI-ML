```python
# Arithmetic Operators:

x = 10; y = 3
sum = x+y
sub = x-y
mul = x*y
div = x/y
rem = x%y
print (sum,sub,mul,div,rem)
```

```python
# Comparison Operators:

x = 10; y = 3
greater_than = x > y
greater_than_equal = x >= y
print(greater_than, greater_than_equal)
```

```python
# Logicaal Operators:

x=10; y=5; z=15
result = (x>y) and (x<z)
print(result)
```

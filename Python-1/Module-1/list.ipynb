```python
# Accessing list elements:

# Indexing:
fruits = ['apple', 'banana', 'mango']
print(fruits[0])
print(fruits[-1])
```

```python
# Slicing:

num = [1,2,3,4,5]
print(num[1:4])
```

```python
# List methods:

# append(): Add item at the end of the list

fruits = ['apple']
fruits.append('orange')
print(fruits)
```

```python
# insert(): Add item at a specific index

fruits = ['apple', 'banana']
fruits.insert(1, 'orange')
print(fruits)
```

```python
# remove():

fruits = ['apple', 'banana', 'cherry']
fruits.remove('cherry')
print(fruits)
```

```python
# pop():

num = [1,2,3,4,5]
num.pop()
print(num)
```

```python
# clear():

nums = [1,2,3,4,5]
nums.clear()
print(nums)
```

```python
# index():

fruits = ['apple', 'banana', 'mango']
print(fruits.index('banana'))
```

```python
# count():
fruits = ['apple', 'banana', 'apple']
print(fruits.count('apple'))
```

```python
# sort():

nums = [3,1,4,2,5]
nums.sort()
print(nums)
```

```python
# reverse():

nums = [1,2,3,4,5]
nums.reverse()
print(nums)
```

```python
# copy():

a = [1,2,3]
b = a.copy()
print(b)
```

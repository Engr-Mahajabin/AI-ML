```python
# Opening a file:

file = open ("./example.txt", "r")
```

```python
# Read a file:

file = open ("./example.txt", "r")
content = file.read()
print(content)
```

```python
# Read line from a file:

file = open ("./example.txt", "r")
content = file.readline()
print(content)
```

```python
# Read lines from a file:

file = open ("./example.txt", "r")
content = file.readlines()
print(content)
```

```python
# Closing a file:

file = open ("./example.txt", "r")
content = file.readlines()
print(content)
print(file.closed)
```

```python
# Using with statement to handle files:

with open("./example.txt", "r") as file:
    content = file.readlines()
    print(content)
print(file.closed)
```

```python
# Write in a file(overwrite):

with open("./example.txt", "w") as file:
    file.write("Hello,Everyone!")
```

```python
# Write in a file without overwriting:

with open("./example.txt", "a") as file:
    file.write("\nWelcome to Phitron!")
```

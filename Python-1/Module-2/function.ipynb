```python
# Function:

def welcome_message():
    print("Welcome to the Python programming world!")

# Call the function
welcome_message()
```

```python
# Function with parameters:

def greet(name):
    print(f"Hello", name)

greet("Ritu")
```

```python
#Default parameters:

def greet(name="None"):
    print("Hello", name)

greet()
    
```

```python
# Function with Return value:

def add(a,b):
    return a+b

result = add(5,5) * 5
print(result)
```

```python
# Return type of a function:

def add(a,b):
    return a+b

result = add(5,5) * 5
print(result)
print(type(result))
```

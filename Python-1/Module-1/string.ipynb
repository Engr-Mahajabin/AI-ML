```python
# Accessing string:
# Indexing:

text = "python"
print(text[0])
```

```python
# Slicing:

text = "phitron"
print(text[0:4])
print(text[0:3])
print(text[2:])
```

```python
# Common string methods:

# upper()/lower():

name = "ritu"
print(name.upper())
print(name.lower())
```

```python
# strip(): Remove spaces

name = "     ritu    "
print(name.strip())
```

```python
# replace():

text = "Hello world"
print(text.replace("world", "Ritu"))
```

```python
# split(): Convert string to list

text = "apple, banana, mango"
print(text.split())
```

```python
# join(): Convert list to string

words = ["I", "love", "Python"]
print(" ".join(words))
```

```python
# find():

text = "Hello"
print(text.find("o"))
print(text.find("p"))
```

```python
# count():

text = "banana"
print(text.count("a"))
```

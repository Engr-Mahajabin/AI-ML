```python
# Problem - 2: 

def compute(a,b):
    try:
        return (a**2 + b**2) / (a - b)
    except ZeroDivisionError:
        return "Error: Division not possible"

a,b = map(int, input().split())
print(compute(a,b))
```

```python
# Problem - 3: Write a Python program that reads a text file and finds the most frequently occurring word:

import string
with open("data.txt", "r") as file:
        data = file.read().lower()

for char in string.punctuation:
    data = data.replace(char, "")

word = data.split()
max_word = max(word, key=word.count)
print(max_word)
```

```python
# Problem - 4: Write a program that converts a sentence into a list representing the length of each word.Ignore extra spaces:

n = str(input()) # Machine learning is fun
words = n.split()
char = list(map(lambda x: len(x), words))
print(char)
```

```python
# Problem - 5: Write a program to check whether a number is a mirror number:

n = int(input()) # 121
x = str(n)
y = x[::-1]
if x == y:
    print("Mirror Number")
else: 
    print("Not Mirror Number")
```

```python
# Positional arguments:

def adder(a,b,c):
    print(f"{a},{b},{c}")

adder(10,20,30)
```

```python
# Arbitrary arguments(*args): Multiple values(returns a tuple)

def total (*num):
    print(num)

total(10,20,30)
```

```python
# Arbitrary keyword arguments(**kwargs): Multiple key-value pairs(returns a dictionary)

def adder(**num):
    print(num)

adder(a=10,b=30,c=50)
```

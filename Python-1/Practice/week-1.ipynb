```python
# Problem 4:
def power(base, exponent=None):
    if exponent is None:
        return base ** 2
    return base ** exponent

print(power(5))
print(power(5,3))
```

```python
# Practice Day 2: https://docs.google.com/document/d/1LtHBJvCbMWHhaH1NbysZF5lg2cNZmSAO/edit

# Problem 1: Create and Write to a File:

def create_file(filename,content):
    with open(filename, "w") as file:
        file.write(content)
create_file("notes.txt", "Hello, this is my first file.")
```

```python
# Problem 2: Read Lines with Line Numbers:

def read_lines(filename):
    with open(filename, "r") as file:
        for index, line in enumerate(file, start=1):
            print(f"{index}: {line.strip()}")
read_lines("data.txt")
```

```python
# Problem 3: Append Log Entries to a File:

def add_log(message):
    with open("log.txt", "a") as file:
        file.write(message + "\n")
# add_log("User logged in")
# add_log("File uploaded")
```

```python
# Problem 4: Read from a Specific File Pointer Position:

def read_from_position(filename, position):
    with open(filename, "r") as file:
        file.seek(position)
        return file.read()
result = read_from_position("hello.txt", 7)
print(result)
```

```python
# Problem 5: Safe File Read with Full Exception Handling:

def safe_read(filename):
  try:
    with open(filename, "r") as file:
        data = file.read()
  except FileNotFoundError:
        print("Error: File not found.")
  else:
      print(data)
  finally: 
      print("Operation complete.")
safe_read("missing.txt")
# safe_read("hello.txt")

```

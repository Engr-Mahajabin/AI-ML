```python
# Iterator:

nums = [1,2,3,4,5]
it = iter(nums) # iter() -> creates iterator
print(next(it)) # next() -> gets next value
print(next(it))
print(next(it))
```

```python
# Generator:

data = [x for x in range(50)]

def data_loader(data, chunk_size):
    for i in range(0, len(data), chunk_size):
        yield data[i:i+chunk_size]

x = data_loader(data, 5)

print(next(x))  # [0, 1, 2, 3, 4]
print(next(x))  # [5, 6, 7, 8, 9]
```

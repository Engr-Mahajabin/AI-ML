```python
# Problem 2:
data = [1, 2, 3, 4, 5, 6, 7, 8]

def batch_generator(batch_size,data):
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]
x = batch_generator(3,data)
print(next(x))
print(next(x))
print(next(x))
```

```python
# Problem 3:
nums = [1,2,3,4]
sqr = list(map(lambda x : x**2,nums))
print(sqr)
```

```python
# Problem 4:
def power(base, exponent=None):
    if exponent is None:
        return base ** 2
    return base ** exponent

print(power(5))
print(power(5,3))
```

```python
# Problem 5:
nums = [1,2,3,4,5,6]
sqr = list(map(lambda x : x**2, filter(lambda x : x%2 == 0, nums)))
print(sqr)
```
